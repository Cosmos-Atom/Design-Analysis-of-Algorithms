Question - 1

```python
def sum_of_natural_iterative(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i
    
    return sum

def sum_of_natural_recursive(n):
    if n == 1:
        return 1
    else:
        return n + sum_of_natural_recursive(n - 1)
    
    
def measure_time(func, n):
    start_time = time.time()
    func(n)
    return time.time() - start_time

N_values = list(range(1, 1001, 1))

iterative_times = [measure_time(sum_of_natural_iterative, n) for n in N_values]
recursive_times = [measure_time(sum_of_natural_recursive, n) for n in N_values]

plt.plot(N_values, iterative_times, label='Iterative')
plt.plot(N_values, recursive_times, label='Recursive')
plt.xlabel('N')
plt.ylabel('Time (s)')
plt.title('Time taken to compute sum of first N natural numbers')
plt.legend()
plt.grid(True)
plt.show()
```

Question - 2


```python
def linear_search(arr, key):
    counter = 0
    for i in arr:
        if i == key:
            counter += 1
    if counter == 0:
        return 'Element not found'
    else:
        return counter

def binary_search(arr, low, high, key):
    if high >= low:
        mid = low + (high - low) // 2
        if arr[mid] == key:
            return 'Element found'
        elif arr[mid] < key:
            return binary_search(arr, mid + 1, high, key)
        else:
            return binary_search(arr, low, mid - 1, key)
    else:
        return 'Element not found'

arr = np.random.randint(low=1, high=1001, size=10000)

search_keys = np.random.randint(low=1, high=1001, size=5)  
linear_times = []
binary_times = []

for key in search_keys:
    # Linear search
    start_time = time.time()
    linear_search(arr, key)
    linear_times.append(time.time() - start_time)
    
    # Binary search
    start_time = time.time()
    binary_search(sorted(arr), 0, len(arr) - 1, key)  
    binary_times.append(time.time() - start_time)

# Plotting
plt.plot(range(1, 6), linear_times, label='Linear Search')
plt.plot(range(1, 6), binary_times, label='Binary Search')
plt.xlabel('Search')
plt.ylabel('Time (s)')
plt.title('Time taken by Linear and Binary Searches')
plt.xticks(range(1, 6))
plt.legend()
plt.grid(True)
plt.show()
```

```python
def string_to_integer(s):
    if len(s) == 0:
        return 0
    else:
        return int(s[0]) + 10 * string_to_integer(s[1:])

input_string = "13531"
result = string_to_integer(input_string)
print(f"The integer representation of '{input_string}' is: {result}")
```

```python
str = 'DAA'
def reverse_string_recursive(str, n):
    if n == 0:
        return ''
    else:
        return str[n - 1] +  reverse_string_recursive(str, n-1)

print(reverse_string_recursive(str, len(str)))
```

```python
def palindrome_check(in_str):
    reversed_str = reverse_string_recursive(in_str, len(in_str))
    if reversed_str == in_str:
        return 'Given string is a palindrome'
    else:
        return ' Given string is not a palindrome'
    
palindrome_check('poop')
```
